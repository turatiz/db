```python
import numpy as np
import matplotlib.pyplot as plt

# Simulación de datos ficticios (meses y uso de recursos en porcentaje o GB)
meses = np.array([1, 2, 3, 4, 5, 6])  # Últimos 6 meses
uso_cpu = np.array([40, 45, 50, 60, 65, 70])  # Uso de CPU en %
uso_memoria = np.array([30, 35, 40, 45, 50, 55])  # Uso de memoria en %
uso_almacenamiento = np.array([2.0,2.5,3.0,3.5,4.0,4.5])  # Almacenamiento en TB

# Aplicamos regresión lineal para encontrar la tendencia
coef_cpu = np.polyfit(meses, uso_cpu, 1)  # Ajuste lineal CPU
coef_mem = np.polyfit(meses, uso_memoria, 1)  # Ajuste lineal Memoria
coef_alma = np.polyfit(meses, uso_almacenamiento, 1)  # Ajuste lineal Almacenamiento

# Función de tendencia
prediccion_meses = np.array([7, 8, 9])  # Próximos 3 meses
pred_cpu = np.polyval(coef_cpu, prediccion_meses)
pred_mem = np.polyval(coef_mem, prediccion_meses)
pred_alma = np.polyval(coef_alma, prediccion_meses)

# Graficamos tendencias
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.plot(meses, uso_cpu, 'bo-', label='Histórico CPU')
plt.plot(prediccion_meses, pred_cpu, 'ro--', label='Predicción CPU')
plt.xlabel('Meses')
plt.ylabel('Uso CPU (%)')
plt.legend()
plt.title('Tendencia Uso CPU')

plt.subplot(1, 3, 2)
plt.plot(meses, uso_memoria, 'go-', label='Histórico Memoria')
plt.plot(prediccion_meses, pred_mem, 'ro--', label='Predicción Memoria')
plt.xlabel('Meses')
plt.ylabel('Uso Memoria (%)')
plt.legend()
plt.title('Tendencia Uso Memoria')

plt.subplot(1, 3, 3)
plt.plot(meses, uso_almacenamiento, 'mo-', label='Histórico Almacenamiento')
plt.plot(prediccion_meses, pred_alma, 'ro--', label='Predicción Almacenamiento')
plt.xlabel('Meses')
plt.ylabel('Almacenamiento (GB)')
plt.legend()
plt.title('Tendencia Uso Almacenamiento')

plt.tight_layout()
plt.show()

# Resultados de la predicción
for i, mes in enumerate(prediccion_meses):
    print(f"Mes {mes}: CPU={pred_cpu[i]:.2f}%, Memoria={pred_mem[i]:.2f}%, Almacenamiento={pred_alma[i]:.2f} GB")
```
